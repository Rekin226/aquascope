# AquaScope Beginner Tutorial

Welcome! This notebook walks you through AquaScope's core features step by step.

**What you'll learn:**
1. Collecting water data from global sources
2. Running exploratory data analysis (EDA)
3. Getting AI methodology recommendations
4. Visualising results

**Prerequisites:** `pip install aquascope[all]`

```python
import aquascope
print(f"AquaScope v{aquascope.__version__}")
```

## 1. Collecting Data

AquaScope has 10 built-in collectors. Let's start with the simplest — Open-Meteo (free, no API key):

```python
from aquascope.collectors import OpenMeteoCollector

meteo = OpenMeteoCollector()
raw = meteo.fetch_raw(lat=25.03, lon=121.57, mode="weather", start_date="2024-01-01", end_date="2024-01-31")
records = meteo.normalise(raw)
print(f"Got {len(records)} records")
records[0] if records else "No records"
```

## 2. Exploratory Data Analysis

Let's save the data and run EDA:

```python
from aquascope.utils.storage import save_records
path = save_records(records, prefix="tutorial", fmt="json")
print(f"Saved to {path}")
```

```python
from aquascope.analysis.eda import run_eda

report = run_eda(str(path))
print(f"Records:    {report.n_records}")
print(f"Stations:   {report.n_stations}")
print(f"Parameters: {report.n_parameters}")
print(f"Date range: {report.date_range}")
```

## 3. AI Methodology Recommendations

```python
from aquascope import recommend

recs = recommend(file=str(path), goal="Analyse weather patterns", top_k=3)
for i, r in enumerate(recs, 1):
    print(f"{i}. {r.methodology.name} (score: {r.score:.2f})")
    print(f"   {r.reasoning}\n")
```

## 4. Visualisation

```python
from aquascope.viz import plot_eda_summary, plot_timeseries
import pandas as pd

# Plot EDA summary
plot_eda_summary(report)
```

```python
# Time-series plot
import pandas as pd, numpy as np
dates = pd.date_range("2024-01-01", periods=31, freq="D")
df = pd.DataFrame({"value": np.random.default_rng(42).normal(15, 5, 31)}, index=dates)
plot_timeseries(df, title="Daily Temperature", ylabel="°C")
```
