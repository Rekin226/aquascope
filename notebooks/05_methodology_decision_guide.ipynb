# Methodology Decision Guide

AquaScope includes 26 built-in research methodologies. This notebook helps you choose the right one.

## Decision Matrix

| Your Goal | Recommended Methods | AquaScope Pipeline |
|-----------|--------------------|--------------------|
| Trend analysis | Mann-Kendall, Sen's Slope | `trend_analysis` |
| Spatial patterns | IDW, Kriging | `spatial_interpolation` |
| Classify water quality | WQI, Clustering | `water_quality_index` |
| Predict time-series | ARIMA, Prophet, LSTM | `forecast` command |
| Assess flood risk | GEV, HEC-RAS routing | `flood-freq`, `solve` |
| Monitor drought | SPI, Palmer Index | `solve "drought..."` |
| Detect anomalies | Isolation Forest, Z-score | `quality` challenge |
| Compare stations | ANOVA, Kruskal-Wallis | `statistical_comparison` |

## Exploring Available Methodologies

```python
from aquascope.ai_engine.knowledge_base import METHODOLOGIES

# Group by category
from collections import Counter
categories = Counter(m.category for m in METHODOLOGIES)
for cat, count in categories.most_common():
    print(f"  {cat}: {count} methods")
```

```python
# List all methodologies
for m in sorted(METHODOLOGIES, key=lambda x: x.category):
    print(f"  [{m.category}] {m.name}")
    print(f"    ID: {m.method_id} | Min records: {m.min_records} | Needs temporal: {m.requires_temporal}")
    print()
```

## AI Recommendation Example

Let the AI choose for you based on your data characteristics:

```python
from aquascope.ai_engine.recommender import MethodologyRecommender
from aquascope.analysis.eda import DatasetProfile

# Scenario: Long-term monitoring dataset
profile = DatasetProfile(
    n_records=10000, n_stations=20, n_parameters=12,
    time_span_years=10.0,
    parameters=["pH", "DO", "BOD5", "turbidity", "conductivity", "water_temp",
                "NH3-N", "COD", "SS", "E_coli", "nitrate", "phosphate"],
    has_temporal_data=True, has_spatial_data=True,
    goal="Comprehensive water quality assessment with trend and spatial analysis",
)

rec = MethodologyRecommender()
results = rec.recommend(profile, top_k=10)
for i, r in enumerate(results, 1):
    print(f"{i:2d}. {r.methodology.name} (score: {r.score:.2f})")
    print(f"    {r.reasoning}\n")
```

## Model Recommendation Matrix

For predictive modelling, AquaScope has a specialized recommender:

```python
from aquascope.ai_engine.model_recommender import ModelRecommender

mr = ModelRecommender()

scenarios = [
    ("flood", "forecast"),
    ("drought", "forecast"),
    ("quality", "anomaly_detection"),
    ("flood", "classification"),
]

for challenge, task in scenarios:
    models = mr.recommend(challenge_type=challenge, task_type=task, top_k=3)
    print(f"  {challenge} / {task}:")
    for m in models:
        print(f"    → {m['model_id']} (score: {m['score']:.2f})")
    print()
```
